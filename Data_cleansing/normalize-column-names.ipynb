```python
import pandas as pd
```

```python
columns_map = {}

pro_data = pd.read_csv("C:\\Users\\1234\\Desktop\\prototype_final_csv.csv")

for key, value in zip(pro_data.columns, pro_data.iloc[1].tolist()):
    columns_map[key] = value

del columns_map['Unnamed: 49']
columns_map
```

```python
data = pd.read_csv("C:\\Users\\1234\\Desktop\\academy_date_parsed.csv")
for key, value in columns_map.items():
    data[value] = data[key]
    del data[key]
    
data.head()
```

```python
to_be_deleted_columns = ['fully null', 'jalali_year', 'jalali_month', 'jalali_week_year', 'jalali_month_year', 'cal_month', 'cal_week_year', 'cal_year', 'not important']
for c in to_be_deleted_columns:
    del data[c]
```
